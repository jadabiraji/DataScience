```python
import pandas as pd
import folium
import numpy as np
import datetime as dt

pd.options.display.max_columns =30

df = pd.read_csv('Los_Angeles_Crime_2020_2023.csv')
df
```

#### Filter for results after 2023 

```python
df = df.loc[df['DATE OCC'].str.contains('2023')]
```

#### Properly format TIME OCC (some values contain 1 digit only --> turn them to 4 digit values )

```python
time_occ = df.loc[df['TIME OCC'] < 1000, 'TIME OCC'].to_list()

i = 0
for time in time_occ:
    time = str(time)
    while len(time) < 4:
        time = "0" + time
    
    time_occ[i] = time
    i+=1

    
df.loc[df['TIME OCC'] < 1000, 'TIME OCC'] = time_occ
time_occ
```

#### Extract date and time from DATE OCC and TIME OCC to create a new date_time column (TimeStamp)

```python
date = df['DATE OCC'].to_list()
time = df['TIME OCC'].to_list()
date_time = []

time_str = [str(x) for x in time]

for i,j in zip(date, time_str):
    date_time.append(i[3:5] + "-" + i[0:2] + "-" + i[6:10] + " " + j[0:2] + ":" + j[2:])

# Create a new column
df.loc[:, 'DateTime'] = date_time

# Tranform the new column to a datetime datatype
df['DateTime'] = pd.to_datetime(df['DateTime'], format='%d-%m-%Y %H:%M')
```

#### Create 3 new columns Date, Time and Day for better Visualization of the dataframe

```python
df.loc[:,'Date'] = [d.date() for d in df['DateTime']]
df.loc[:, 'Time'] = [d.time() for d in df['DateTime']]
df.loc[:, 'Day'] = df['DateTime'].dt.day_name()
```

```python
df.info()
```

#### Rearrange the order of the columns and keep only the valuable ones

```python
cols = df.columns.to_list()

crime_df = df[ [cols[0]] + [cols[28]]  + [cols[9]] + [cols[24]]+ cols[29:] + [cols[15]] + cols[26:28] ]
crime_df
```

#### Rename Columns, sort the values by DateTime, and set the index as 'Crime ID'

```python
crime_df = crime_df.rename(columns={'DR_NO': 'Crime ID', 'Crm Cd Desc': 'Crime Type' , 'LOCATION' : 'Address', 
                                   'LAT': 'Latitude', 'LON': 'Longitude'})

crime_df = crime_df.sort_values(by=['DateTime'])

crime_df = crime_df.set_index('Crime ID')
```

#### Filtering the categories of crimes

```python
crime_type = crime_df.groupby('Crime Type')['Crime Type'].count().sort_values().to_frame()
for index, row in crime_type.iterrows():
    print(index + ", Count: " + str(row['Crime Type']))
```

#### Save the crime_df to total_crime.csv

```python
crime_df.to_csv('total_crimes_2023.csv')
```

### Creating 3 new dataframe to filter crimes by type
#### 1) crimes related to shooting a firearm
#### 2) cimes related to srious theft or burglary
#### 3) crimes related to humanitarian ethics

```python
# Crimes related to weapons
shooting_mask = ( ( crime_df['Crime Type'] == 'SHOTS FIRED AT MOVING VEHICLE, TRAIN OR AIRCRAF') |
         ( crime_df['Crime Type'] == 'SHOTS FIRED AT INHABITED DWELLING') |
         ( crime_df['Crime Type'] == 'DISCHARGE FIREARMS/SHOTS FIRED') |
         ( crime_df['Crime Type'] == 'SHOTS FIRED AT MOVING VEHICLE, TRAIN OR AIRCRAF')  )

shooting_crime_df = crime_df[shooting_mask]
shooting_crime_df
```

#### Save the filtered dataframe 

```python
shooting_crime_df.to_csv('shooting_crime.csv')
```

```python
# Crimes related to theft
theft_mask = ( ( crime_df['Crime Type'] == 'VEHICLE - STOLEN' ) |
         ( crime_df['Crime Type'] == 'BURGLARY') |
         ( crime_df['Crime Type'] == 'BURGLARY FROM VEHICLE') |
         ( crime_df['Crime Type'] == 'THEFT FROM MOTOR VEHICLE - GRAND ($950.01 AND OVER)') |
         ( crime_df['Crime Type'] == 'ROBBERY') |
         ( crime_df['Crime Type'] == 'THEFT-GRAND ($950.01 & OVER)EXCPT,GUNS,FOWL,LIVESTK,PROD')  )

theft_crime_df = crime_df[theft_mask]
theft_crime_df
```

```python
theft_crime_df.to_csv('theft_crime.csv')
```

```python
# Crimes related to human ethics
humanitarian_crime_mask = ( ( crime_df['Crime Type'] == 'HUMAN TRAFFICKING - INVOLUNTARY SERVITUDE' ) |
         ( crime_df['Crime Type'] == 'CHILD PORNOGRAPHY') |
         ( crime_df['Crime Type'] == 'HUMAN TRAFFICKING - COMMERCIAL SEX ACTS') |
         ( crime_df['Crime Type'] == 'KIDNAPPING - GRAND ATTEMPT') |
         ( crime_df['Crime Type'] == 'RAPE, ATTEMPTED') |
         ( crime_df['Crime Type'] == 'KIDNAPPING') |
         ( crime_df['Crime Type'] == 'RAPE, FORCIBLE')  )

humanitarian_crime_df = crime_df[humanitarian_crime_mask]
humanitarian_crime_df
```

```python
humanitarian_crime_df.to_csv('humanitarian_crime.csv')
```
